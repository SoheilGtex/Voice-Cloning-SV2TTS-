# Voice Cloning Demo (SV2TTS)
This notebook shows how to use the CLI from Python and call the FastAPI.


```python
from pathlib import Path
from voice_cloner.utils.audio import load_wav_mono, ensure_16k_mono
from voice_cloner.backends.coqui_wrapper import CoquiBackend

wav, sr = load_wav_mono(Path('assets/samples/spk.wav'))
wav16 = ensure_16k_mono(wav, sr)
be = CoquiBackend(model_name='tts_models/en/vctk/vits')
y = be.clone_to_waveform(wav16, 'Hello from notebook demo.')
len(y)
```
